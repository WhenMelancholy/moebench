```python
import torch

logits = torch.load("output/logits/baseline.pt")
new_logits = logits
print("new logits shape", new_logits.shape)
prune_list = []
for i in range(new_logits.shape[0]):
    layer_logits = new_logits[i]
    selected_experts = torch.topk(layer_logits, k=8, dim=-1).indices
    print("selected_experts", selected_experts.shape)
    expert_frequency = torch.bincount(selected_experts.flatten(), minlength=64)
    print("expert_frequency", expert_frequency)
    expert_frequency = torch.bincount(selected_experts.flatten(), minlength=64)
    print("expert_frequency.shape", expert_frequency.shape)
    prune_expert = torch.topk(expert_frequency, k=8).indices
    print("prune_expert", prune_expert)
    prune_list.append(prune_expert)
print(prune_list)
torch.save(prune_list, "output/logits/prune_list_baseline.pt")
```
